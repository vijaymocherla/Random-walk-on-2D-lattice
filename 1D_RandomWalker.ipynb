# 1D Random Walk

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import collections
def Walk(nSteps):
    pathChoices = np.random.choice([-1,1],size=nSteps)
    walkerLoc = sum(pathChoices) #final position of the Walker 
    return(walkerLoc)

def myExperiment(nSteps,runs):
    distribution = []
    for i in range(runs):
        distribution.append(Walk(nSteps))
    collectPaths = dict(collections.Counter(distribution))
    probabilityDistribution = {k: v /(total) for total in (sum(collectPaths.values()),) for k, v in collectPaths.items()}
    return distribution, probabilityDistribution

def plotWalker(probabilityDistribution,title='',textstr=''):
    x,y = zip(*sorted(probabilityDistribution.items()))
    plt.plot(x,y)
    plt.xlabel('x',fontsize=14)
    plt.ylabel('F(x)',fontsize=14)
    plt.title(title)
    plt.gcf().text(0.0, -0.1, textstr, fontsize=14)
    plt.savefig(title+'.png',dpi=200)
    plt.show()
    return(x,y)

```

```python
runs = 100000
nSteps= 100
title = 'nSteps'+str(nSteps)+'_runs'+str(runs)
dist, probDist = myExperiment(nSteps,runs)
dist.sort()
x,y = plotWalker(probDist,title)
f = scipy.interpolate.interp1d(x,y)
a = np.linspace(-40,40,1000)  
print(scipy.stats.kstest(f(a),'norm',N=100))
#plt.plot(a,f(a))
print('Variance:',np.var(dist))
print('Standard Deviation:',np.std(dist))
```
