# 1D Random Walk

## Code
Lets write some code here for a random walk on 1-dimensional line of integers.

```python
# Importing a few packages!
import numpy as np
import scipy
import matplotlib.pyplot as plt
import collections

def Walk(nSteps): 
    # This function returns the position of the walker after nSteps
    pathChoices = np.random.choice([-1,1],size=nSteps) # an array of 'n' random choices of -1 or +1 
    walkerLoc = sum(pathChoices) # final position of the walker 
    return(walkerLoc)

def myExperiment(nSteps,runs):
    outcomes = []
    # In the following loop we run multiple trajectories to generate
    # a probability distribution of position of the walker after nSteps. 
    for i in range(runs):
        outcomes.append(Walk(nSteps))
    collectPaths = dict(collections.Counter(outcomes)) # Here we count the frequency of outcomes 
    probabilityDistribution = {k: v /(total) for total in (sum(collectPaths.values()),) 
                               for k, v in collectPaths.items()}
    return(outcomes,probabilityDistribution)

def plotWalker(probabilityDistribution,title='',textstr=''):
    # Plotting the probability distribution 
    x,y = zip(*sorted(probabilityDistribution.items()))
    plt.plot(x,y)
    plt.xlabel('x',fontsize=14)
    plt.ylabel('F(x)',fontsize=14)
    plt.title(title)
    plt.gcf().text(0.0, -0.1, textstr, fontsize=14)
    plt.savefig(title+'.png',dpi=200)
    plt.show()
    return(x,y)

```

### Lets Random walk!

The probability to find the random walker at a certain point in the lattice after 'N' steps takes the form of a gaussian distribution. We shall see this in the following cells. 

```python
runs = 100000
nSteps= 100
title_one = 'nSteps'+str(nSteps)+'_runs'+str(runs)
outcomes_one,probDist_one = myExperiment(nSteps,runs)
x1,y1 = plotWalker(probDist_one,title_one)
f = scipy.interpolate.interp1d(x1,y1)
a = np.linspace(-40,40,1000)  
print(scipy.stats.kstest(f(a),'norm',N=100))
#plt.plot(a,f(a))
print('Variance:',np.var(outcomes_one))
print('Standard Deviation:',np.std(outcomes_one))
```
